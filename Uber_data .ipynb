```python
# remove all rows where total amount and number of passangers was zero

df_cleaned = df[~(
    (df['trip_distance'] == 0) |
    (df['passenger_count'] == 0) |
    (df['total_amount']==0) |
    (df['pickup_longitude']==0) |
    (df['pickup_latitude']==0) |
    (df['dropoff_longitude']==0) |
    (df['dropoff_latitude']==0)
    )].reset_index(drop=True)

df_cleaned
```

```python
# Remove duplicates and reset indexes

df_cleaned = df_cleaned.drop_duplicates().reset_index(drop=True)

df_cleaned
```

```python
# remove all rows with total_amount superior to 300 or trip distance greater than 50 km
# there is only 4 trips with a distance greater then 50 km and 5 trips with a total amount greater than 300$

df_cleaned = df[~(
    (df['total_amount'] > 300) |
    (df['trip_distance'] > 50)
    )].reset_index(drop=True)

df_cleaned
```

```python
df_cleaned['trip_id'] = df_cleaned.index
df_cleaned.rename(columns={'VendorID': 'vendorid'}, inplace=True)
df_cleaned.rename(columns={'RatecodeID': 'ratecodeid'}, inplace=True)
```

##Table Creation.

According to our modelling, we have to create 9 differents tables.
- Vendor
- Payment


```python
# Vendor Table
# This dimension table contains the list of vendors and their information. We could have name, plate , car ...

vendor_dict = {'vendorid': list(set(df_cleaned['vendorid']))}
Vendor = pd.DataFrame(vendor_dict)

Vendor
```

```python
#Payment Table: It is a fact table that contains for each trip the payment corresponding

Payment = df_cleaned[['trip_id','fare_amount','extra','mta_tax','tip_amount','tolls_amount','improvement_surcharge','total_amount']]

Payment.rename(columns={'trip_id': 'payment_id'}, inplace=True)

Payment
```

```python
# This table contains all drop locations.

Dropoff_Location = df_cleaned[['trip_id','dropoff_latitude','dropoff_longitude']]

Dropoff_Location.rename(columns={'trip_id': 'dropoff_id'}, inplace=True)

Dropoff_Location
```

```python
# This Table contains all pick up locations and ID

Pickup_Location = df_cleaned[['trip_id','pickup_latitude','pickup_longitude']]

Pickup_Location.rename(columns={'trip_id': 'pickup_id'}, inplace=True)

Pickup_Location
```

```python
# Ratecode Dimension Table

Ratecode_dict = {
    1:"Standard rate",
    2:"JFK",
    3:"Newark",
    4:"Nassau or Westchester",
    5:"Negotiated fare",
    6:"Group ride"
}

Ratecode_dict = {
    'ratecodeid': list(Ratecode_dict.keys()),
    'rate_code_name': list(Ratecode_dict.values())
}
Ratecode = pd.DataFrame(Ratecode_dict)

Ratecode
```

```python
# Payment type Dimension Table

payment_type_name = {
    1:"Credit card",
    2:"Cash",
    3:"No charge",
    4:"Dispute",
    5:"Unknown",
    6:"Voided trip"
}

payment_type_dict = {
    'payment_type_id' : list(payment_type_name.keys()),
    'payment_type_name' : list(payment_type_name.values())
}

payment_type = pd.DataFrame(payment_type_dict)

payment_type
```

```python
# This Table contains all pick up locations and ID

Dropoff_time = df_cleaned[['trip_id','tpep_dropoff_datetime']]

Dropoff_time.rename(columns={'trip_id': 'dropoff_time_id'}, inplace=True)

Dropoff_time
```

```python
# This Table contains all pick up locations and ID

Pickup_time = df_cleaned[['trip_id','tpep_pickup_datetime']]

Pickup_time.rename(columns={'trip_id': 'pickup_time_id'}, inplace=True)

Pickup_time
```

```python
Trip = df_cleaned.merge(Payment, left_on='trip_id', right_on='payment_id')\
             .merge(Dropoff_Location, left_on='trip_id', right_on='dropoff_id') \
             .merge(Pickup_Location, left_on='trip_id', right_on='pickup_id') \
             .merge(Ratecode, left_on='ratecodeid', right_on='ratecodeid') \
             .merge(payment_type, left_on='trip_id', right_on='payment_type_id')\
             .merge(Dropoff_time, left_on='trip_id', right_on='dropoff_time_id') \
             .merge(Pickup_time, left_on='trip_id', right_on='pickup_time_id') \
             .merge(Vendor, left_on='vendorid', right_on='vendorid') \
             [['trip_id','vendorid', 'pickup_time_id', 'dropoff_time_id',
               'payment_type_id', 'payment_id', 'pickup_id', 'dropoff_id','ratecodeid','trip_distance','store_and_fwd_flag','passenger_count']]
```
