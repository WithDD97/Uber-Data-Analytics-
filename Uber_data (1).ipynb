```python
# remove all rows where total amount and number of passangers was zero

df_cleaned = df[~(df['total_amount']==0)]

df_cleaned = df_cleaned[~(df_cleaned['passenger_count']==0)]

df_cleaned
```

```python
# Remove duplicates and reset indexes

df_cleaned = df_cleaned.drop_duplicates().reset_index(drop=True)

df_cleaned
```

```python
df_cleaned['trip_ID'] = df_cleaned.index


```

```python
# Vendor Table

vendor_dict = {'VendorID': list(set(df_cleaned['VendorID']))}
Vendor = pd.DataFrame(vendor_dict)

Vendor
```

```python
#Payment Table

Payment = df_cleaned[['trip_ID','fare_amount','extra','mta_tax','tip_amount','tolls_amount','improvement_surcharge','total_amount']]

Payment
```

```python
Dropoff_Location = df_cleaned[['dropoff_latitude','dropoff_longitude']].drop_duplicates().reset_index(drop=True)

Dropoff_Location['dropoff_ID'] = Dropoff_Location.index

Dropoff_Location
```

```python
Pickup_Location = df_cleaned[['pickup_latitude','pickup_longitude']].drop_duplicates().reset_index(drop=True)

Pickup_Location['pickup_ID'] = Pickup_Location.index

Pickup_Location
```

```python
Ratecode_dict = {
    1:"Standard rate",
    2:"JFK",
    3:"Newark",
    4:"Nassau or Westchester",
    5:"Negotiated fare",
    6:"Group ride"
}

Ratecode_dict = {
    'RatecodeID': list(Ratecode_dict.keys()),
    'rate_code_name': list(Ratecode_dict.values())
}
Ratecode = pd.DataFrame(Ratecode_dict, )

Ratecode
```
